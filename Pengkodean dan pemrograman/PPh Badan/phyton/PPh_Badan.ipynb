## Data cleaning

### Subtask:
Clean the three dataframes: `df_aset`, `df_kebijakan`, and `df_transaksi`.


**Reasoning**:

Outlier management will be done using winsorization, as it's a robust method that doesn't remove data points.

## Data wrangling

### Subtask:
Combine the three cleaned DataFrames (`df_aset`, `df_kebijakan`, and `df_transaksi`) into a single DataFrame for analysis.  Create new features if necessary.


**Reasoning**:
Combine the three dataframes based on the common column 'tahun' and create new features.



```python
# Merge df_aset and df_transaksi.  Since there is no common column between these two dataframes,
# we cannot merge them based on a common column.  We'll need to add a common column.
# Assuming 'tahun' is related to the transactions, we'll add a 'tahun' column to df_aset.
# We'll arbitrarily assign the same year to all assets in df_aset for demonstration purposes.
# In a real-world scenario, you would need to determine the appropriate year for each asset.
df_aset['tahun'] = 2023  # Replace with the appropriate year for each asset

# Merge df_aset and df_transaksi
df_merged = pd.merge(df_aset, df_transaksi, on='tahun', how='inner')

# Merge the result with df_kebijakan
df_final = pd.merge(df_merged, df_kebijakan, on='tahun', how='inner')

# Feature Engineering
# Calculate Depreciation Expense
df_final['depresiasi_hitung'] = df_final['nilai_perolehan'] / df_final['umur_ekonomis']

# Calculate Net Profit Margin
df_final['laba_bersih'] = df_final['pendapatan'] - df_final['beban_operasional'] - df_final['penyusutan']
df_final['net_profit_margin'] = (df_final['laba_bersih'] / df_final['pendapatan']) * 100

# Calculate Effective Tax Rate
df_final['tax_holiday_flag'] = df_final.apply(lambda row: 1 if row['tahun'] >= row['tax_holiday_awal'] and row['tahun'] <= row['tax_holiday_akhir'] else 0, axis=1)

df_final['pajak'] = df_final['laba_bersih'] * df_final['tax_rate']
df_final['effective_tax_rate'] = (df_final['pajak'] / df_final['laba_bersih']) * 100

# Handle potential missing values after merge and feature engineering
# Impute missing values using mean, median or other methods.
for col in df_final.columns:
    if df_final[col].isnull().any():
        if pd.api.types.is_numeric_dtype(df_final[col]):
            df_final[col].fillna(df_final[col].median(), inplace=True)
        else:
            df_final[col].fillna(df_final[col].mode()[0], inplace=True)

display(df_final)
```

## Data analysis

### Subtask:
Perform descriptive statistics and identify correlations within the combined dataset (`df_final`)


**Reasoning**:
Calculate descriptive statistics for numerical columns and identify correlations between variables in df_final. Visualize the correlation matrix using a heatmap. Segment the data based on 'kategori' and perform the same analysis for each segment. Summarize the findings.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Descriptive Statistics
numerical_cols = df_final.select_dtypes(include=['number']).columns
descriptive_stats = df_final[numerical_cols].describe()
display(descriptive_stats)

# Correlation Matrix and Heatmap
correlation_matrix = df_final[numerical_cols].corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()


# Segmentation and Analysis by 'kategori'
segments = df_final['kategori'].unique()
for segment in segments:
    segment_data = df_final[df_final['kategori'] == segment]
    print(f"\nDescriptive Statistics for Segment: {segment}")
    display(segment_data[numerical_cols].describe())

    print(f"\nCorrelation Matrix for Segment: {segment}")
    segment_correlation_matrix = segment_data[numerical_cols].corr()
    plt.figure(figsize=(10, 8))
    sns.heatmap(segment_correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title(f'Correlation Matrix for Segment: {segment}')
    plt.show()

# Summary of Findings
print("\nSummary of Findings:")
print("The descriptive statistics provide insights into the distribution of numerical variables, including mean, median, standard deviation, min, max, and quartiles. The newly engineered features 'depresiasi_hitung', 'net_profit_margin', and 'effective_tax_rate' can be examined for their statistical properties.")
print("The correlation matrix and heatmap reveals relationships between numerical variables. Strong positive or negative correlations can be identified, indicating potential dependencies between variables. The focus should be on correlations involving the newly engineered features to understand their relationships with other financial metrics.")
print("Segmenting the data by 'kategori' allows for a more granular analysis of how the relationships between variables vary across different asset categories.  Descriptive statistics and correlation matrices for each segment provide insights into category-specific trends.")
```
